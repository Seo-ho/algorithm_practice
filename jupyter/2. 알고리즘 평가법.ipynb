## 시간 복잡도
- 데이터가 많아질수록 걸리는 시간이 얼마나 급격히 증가하는가


### 점근 표기법(Big-O Notation)
- n이 크다고 가정. n은 input의 크기
- 절대적인 시간이 아닌 성장성을 보기에. 



## 공간 복잡도
- input_size에 비례해서 알고리즘이 사용하는 메모리 공간을 나타냄.
- 시간 복잡도도 그렇지만 input_size와 무관하면 O(1)

```python
my_list = [7, 5, 2, 3, 6]

# 시간복잡도: O(nlogn)
print(sorted(my_list))       # => [2, 3, 5, 6, 7]
print(my_list)               # => [7, 5, 2, 3, 6]

my_list.sort()
print(my_list)               # => [2, 3, 5, 6, 7]
```
